Для **DataFrame iris** рассчитайте следующие виды метрик:


#### <font color='darkred'>Центральные метрики:</font>

* Среднеарифметическое;
* Средневзвешенное;
* Среднегеометрическое;
* Гармоническое среднее;
* Медиана;
* Мода.


#### <font color='darkred'>Метрики оценки вариативности данных:</font>
* Дисперсия;
* СКО;
* Смещение;
* Процентили;
* Диапазон

#### <font color='darkred'>Метрики оценки корреляции или взаимозависимости:</font>
* Ковариация;
* Коэффициент корреляции.

<br>

**_Код скрипт файла_**:


```python
import scipy.stats
import numpy as np
import pandas as pd
import statistics as stat
import sklearn
from sklearn import datasets

iris = datasets.load_iris()
iris_frame = pd.DataFrame(iris.data)
iris_frame.columns = iris.feature_names
iris_frame['target'] = iris.target
iris_frame['name'] = iris_frame.target.apply(lambda x: iris.target_names[x])

for target_name in set(iris_frame['name']):
    print('Цветок - ' + target_name, end='\n')
    for feature_name in iris_frame.columns[: - 2]:
        local_data = iris_frame[iris_frame.name == target_name][feature_name]
        print('Среднеарифмитическое ' + feature_name + ': ' + str(stat.mean(local_data)), end='\n')
        print('Средневзвешенное ' + feature_name + ': ' + str(np.average(local_data)), end='\n')
        print('Среднегеометрическое ' + feature_name + ': ' + str(stat.geometric_mean(local_data)), end='\n')
        print('Гармоническое среднее ' + feature_name + ': ' + str(stat.harmonic_mean(local_data)), end='\n')
        print('Медиана ' + feature_name + ': ' + str(stat.median(local_data)), end='\n')
        print('Мода ' + feature_name + ': ' + str(stat.mode(local_data)), end='\n')
        print('Дисперсия ' + feature_name + ': ' + str(stat.variance(local_data)), end='\n')
        print('СКО ' + feature_name + ': ' + str(stat.stdev(local_data)), end='\n')
        print('Смещение ' + feature_name + ': ' + str(scipy.stats.skew(local_data)), end='\n')
        print('Процентили ' + feature_name + ': ' + str(np.percentile(local_data, 1)), end='\n')
        print('Диапазон ' + feature_name + ': ' + str(np.amax(local_data) - np.amin(local_data)), end='\n')
        print()

flowers_list = []
for flower_name in set(iris_frame['name']):
    flowers_list.append(flower_name)
for i, flower_1 in enumerate(flowers_list):
    for flower_2 in flowers_list[i + 1:]:
        print('Взаимосвязь соответствующих параметров у цветков ' + flower_1 + ' и ' + flower_2, end='\n')
        print()
        print('Коэффициенты корреляции для: ', end='\n')
        for feature_name in iris_frame.columns[: - 2]:
            print(feature_name[:-5] + ': ' + str(
                scipy.stats.pearsonr(iris_frame[iris_frame.name == flower_1][feature_name],
                                     iris_frame[iris_frame.name == flower_2][feature_name])), end='\n')
        print('Ковариация для: ', end='\n')
        for feature_name in iris_frame.columns[: - 2]:
            print(feature_name[:-4] + ': ' + str(np.cov(iris_frame[iris_frame.name == flower_1][feature_name],
                                                                 iris_frame[iris_frame.name == flower_2][
                                                                     feature_name])))
```
